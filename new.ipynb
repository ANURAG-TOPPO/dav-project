```python
import pandas as pd
```

```python
df= pd.read_csv(r'C:\Users\prade\dav proj\data1.csv')
df.drop(columns=['URLs','Unnamed: 4','Unnamed: 5'],inplace=True)
df["Body"] = df["Body"].astype(str)
df["Headline"] = df["Headline"].astype(str)
df.head()
df.shape
```

```python
df['Body'] = df['Body'].str.replace(r'[^\w\s]+', '')
df['Headline'] = df['Headline'].str.replace(r'[^\w\s]+', '')
#convert to lower case
df['Body'] =df['Body'].str.lower()
df['Headline'] = df['Headline'].str.lower()

df.head()
```

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["headline_wo_stop"] = df["Headline"].apply(lambda text: remove_stopwords(text))
df["Body_wo_stop"] = df["Body"].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
import nltk.stem
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["headline_stemmed"] = df["headline_wo_stop"].apply(lambda text: stem_words(text))
df["body_stemmed"] = df["Body_wo_stop"].apply(lambda text: stem_words(text))
df.head()
```

```python
import nltk.tokenize
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
def generate_N_grams(text,ngram):
  tokens = nltk.word_tokenize(text)
  temp=zip(*[tokens[i:] for i in range(0,ngram)])
  ans=[' '.join(ngram) for ngram in temp]
  return ans
df["headline_1g"] = df["headline_stemmed"].apply(lambda text: generate_N_grams(text,1))
df["body_1g"] = df["body_stemmed"].apply(lambda text: generate_N_grams(text,1))
df["headline_2g"] = df["headline_stemmed"].apply(lambda text: generate_N_grams(text,2))
df["body_2g"] = df["body_stemmed"].apply(lambda text: generate_N_grams(text,2))
df["headline_3g"] = df["headline_stemmed"].apply(lambda text: generate_N_grams(text,3))
df["body_3g"] = df["body_stemmed"].apply(lambda text: generate_N_grams(text,3))
df["headline_4g"] = df["headline_stemmed"].apply(lambda text: generate_N_grams(text,4))
df["body_4g"] = df["body_stemmed"].apply(lambda text: generate_N_grams(text,4))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tr_idf_model  = TfidfVectorizer(ngram_range=(2,2))

tf_idf_vector = tr_idf_model.fit_transform(df['body_stemmed'][0:10])
w=tr_idf_model.get_feature_names()


tf_idf_array = tf_idf_vector.toarray()
df_tf_idf = pd.DataFrame(tf_idf_array, columns = w)

df_tf_idf
```
